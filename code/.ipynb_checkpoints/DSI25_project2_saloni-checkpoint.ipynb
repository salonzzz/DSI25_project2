## 1. Importing libraries

## a) Dealing with missing values

* Cross checked with [data description](http://jse.amstat.org/v19n3/decock/DataDocumentation.txt) and converted numerical features to categorical features
* Converted "Na" values to "None" where "Na" is a class i.e. ordinal/nominal
* Dropped columns where >5% of data was missing and also dropped the "Id" column as it is not a predictor of sale price
* Dropped rows with NA where it represents <5% of data

```python
#Check percentage of missing values in each column and dtype of each column
#Cross check with data description provided (http://jse.amstat.org/v19n3/decock/DataDocumentation.txt)
percent_missing = ames_housing.isnull().sum() / len(ames_housing) * 100
missing_value_ames = pd.DataFrame({"dtype":ames_housing.dtypes,
                                   "percent_missing": percent_missing})
print(missing_value_ames)
```

```python
# Some features present as numeric but are actually ordinal/nominal/date a.k.a. categorical
# and need to be converted to str format 
# Reference 1: https://stackoverflow.com/questions/50847374/convert-multiple-columns-to-string-in-pandas-dataframe/50847580
# Reference 2: https://stackoverflow.com/questions/41514173/change-multiple-columns-in-pandas-dataframe-to-datetime

#Convert numerics to string
ames_housing["PID"] = ames_housing["PID"].astype(str)
ames_housing["MS SubClass"] = ames_housing["MS SubClass"].astype(str) 
ames_housing["Overall Qual"] = ames_housing["Overall Qual"].astype(str) 
ames_housing["Overall Cond"] = ames_housing["Overall Cond"].astype(str) 
ames_housing["Bsmt Full Bath"] = ames_housing["Bsmt Full Bath"].astype(str) 
ames_housing["Bsmt Half Bath"] = ames_housing["Bsmt Half Bath"].astype(str) 
ames_housing["Full Bath"] = ames_housing["Full Bath"].astype(str) 
ames_housing["Half Bath"] = ames_housing["Half Bath"].astype(str) 
ames_housing["Bedroom AbvGr"] = ames_housing["Bedroom AbvGr"].astype(str) 
ames_housing["Kitchen AbvGr"] = ames_housing["Kitchen AbvGr"].astype(str) 
ames_housing["TotRms AbvGrd"] = ames_housing["TotRms AbvGrd"].astype(str) 
ames_housing["Fireplaces"] = ames_housing["Fireplaces"].astype(str) 
ames_housing["Garage Cars"] = ames_housing["Garage Cars"].astype(str) 


#Convert dates to str, not using datetime because it isn't accurate
ames_housing["Year Built"] = ames_housing["Year Built"].astype(str)
ames_housing["Year Remod/Add"] = ames_housing["Year Remod/Add"].astype(str)
ames_housing["Garage Yr Blt"] = ames_housing["Garage Yr Blt"].astype(str)
ames_housing["Yr Sold"] = ames_housing["Yr Sold"].astype(str)
ames_housing["Mo Sold"] = ames_housing["Mo Sold"].astype(str)
```

```python
#Change NA to “None” for selected features because NA represents a class

ames_housing["Alley"] = ames_housing["Alley"].fillna("None")
ames_housing["Bsmt Qual"] = ames_housing["Bsmt Qual"].fillna("None")
ames_housing["Bsmt Cond"] = ames_housing["Bsmt Cond"].fillna("None")
ames_housing["Bsmt Exposure"] = ames_housing["Bsmt Exposure"].fillna("None")
ames_housing["BsmtFin Type 1"] = ames_housing["BsmtFin Type 1"].fillna("None")
ames_housing["BsmtFin Type 2"] = ames_housing["BsmtFin Type 2"].fillna("None")
ames_housing["Fireplace Qu"] = ames_housing["Fireplace Qu"].fillna("None")
ames_housing["Garage Type"] = ames_housing["Garage Type"].fillna("None")
ames_housing["Garage Finish"] = ames_housing["Garage Finish"].fillna("None")
ames_housing["Garage Qual"] = ames_housing["Garage Qual"].fillna("None")
ames_housing["Garage Cond"] = ames_housing["Garage Cond"].fillna("None")
ames_housing["Pool QC"] = ames_housing["Pool QC"].fillna("None")
ames_housing["Fence"] = ames_housing["Fence"].fillna("None")
ames_housing["Misc Feature"] = ames_housing["Misc Feature"].fillna("None")
```

```python
#Drop columns where >5% of data is missing + drop "ID" column as it is not a predictor for sales price
ames_housing.drop(["Lot Frontage","Garage Yr Blt","Id"], axis=1, inplace=True)
```

```python
#Drop rows with NA as it represents <5% of data
ames_housing = ames_housing.dropna(subset=["Mas Vnr Type"])
ames_housing = ames_housing.dropna(subset=["BsmtFin SF 1"])
ames_housing = ames_housing.dropna(subset=["BsmtFin SF 2"])
ames_housing = ames_housing.dropna(subset=["Bsmt Unf SF"])
ames_housing = ames_housing.dropna(subset=["Total Bsmt SF"])
ames_housing = ames_housing.dropna(subset=["Bsmt Full Bath"])
ames_housing = ames_housing.dropna(subset=["Bsmt Half Bath"])
ames_housing = ames_housing.dropna(subset=["Garage Cars"])
ames_housing = ames_housing.dropna(subset=["Garage Area"])
```

```python
#Check percentage of missing values in each column and dtype of each column
percent_missing = ames_housing.isnull().sum() / len(ames_housing) * 100
missing_value_ames = pd.DataFrame({"dtype":ames_housing.dtypes,
                                   "percent_missing": percent_missing})
print(missing_value_ames)
ames_housing.shape
```

## b) Dealing with outliers

```python
corr = ames_housing.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
                                  
fig = plt.figure(figsize=(12,12))
ax = sns.heatmap(corr,annot=True,mask = mask, cmap="RdBu",fmt=".2f",
                 vmin = -1, vmax = 1,center=0)
ax.set_title("Correlation Heatmap for Ames Housing Dataset");
```
