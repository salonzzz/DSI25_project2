### 1. Importing libraries

```python
#Data processing and visualization
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import skew

#Modelling
from sklearn.linear_model import LinearRegression, Lasso, Ridge, LassoCV, RidgeCV
from sklearn.preprocessing import OneHotEncoder, PolynomialFeatures, StandardScaler


#Data splitting and evaluation
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.metrics import mean_squared_error, r2_score

#Loading data
import os

#Ignore warnings
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

### 2. Importing dataset

```python
# Import Ames housing data
datapath = "../data"

train_path = os.path.join(datapath,"ames_rem.csv")

train = pd.read_csv(train_path)
```

```python
#Display all columns
pd.set_option("display.max_columns", None)
#Display all rows
pd.set_option("display.max_rows", None)
#View first 5 rows 
train.head()
```

```python
train.shape
```

### 3. Model Prep

```python
#Creating x and y variables
features = [col for col in train._get_numeric_data().columns if col != "SalePrice"]
X = train[features]
y = train["SalePrice"]
```

```python
#train/test/split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### 4. Cross validation

```python
#Models
lr = LinearRegression()
ridge = Ridge()
lasso = Lasso()
```

```python
#Train
lr.fit(X_train, y_train);
ridge.fit(X_train, y_train);
lasso.fit(X_train, y_train);

print(cross_val_score(lr, X_train, y_train, cv=5).mean())
print(cross_val_score(ridge, X_train, y_train, cv=5).mean())
print(cross_val_score(lasso, X_train, y_train, cv=5).mean())
```

```python
#Test
lr.fit(X_test, y_test);
ridge.fit(X_test, y_test);
lasso.fit(X_test, y_test);

print(cross_val_score(lr, X_test, y_test, cv=5).mean())
print(cross_val_score(ridge, X_test, y_test, cv=5).mean())
print(cross_val_score(lasso, X_test, y_test, cv=5).mean())
```

**Notes**

Ridge seems to be the best performing model on both the train and test sets. 

```python
#Ridge is the best model so let's look at it further
ridge = Ridge(alpha=20)
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)

# R2
print(f"R2 score: {r2_score(y_test, y_pred_ridge)}")
# RMSE
print(f"RMSE score: {mean_squared_error(y_test, y_pred_ridge, squared=False)}")
# coefficient
ridge_coefficients = pd.DataFrame()
ridge_coefficients["features"] = X_train.columns
ridge_coefficients["coefficient values"] = ridge.coef_
```

```python
#Scatter plot of train data predicted Sale price vs. Actual Sale price
plt.figure(figsize=(10,6))
sns.scatterplot(y_train, ridge.predict(X_train))
plt.title("Predicted Sale Price vs Actual Sale Price", fontsize=12)
plt.xlabel("Actual", labelpad=15, fontsize=12)
plt.ylabel("Predicted", labelpad=20, rotation =0, fontsize=12);
```

```python
#Scatter plot of test data predicted Sale price vs. Actual Sale price
plt.figure(figsize=(10,6))
sns.scatterplot(y_test, y_pred_ridge)
plt.title("Predicted Sale Price vs Actual Sale Price", fontsize=12)
plt.xlabel("Actual", labelpad=15, fontsize=12)
plt.ylabel("Predicted", labelpad=20, rotation =0, fontsize=12);
```

```python
#Residuals vs. Predicted sale price for train data
resids = ridge.predict(X_train)-y_train

fig,ax=plt.subplots(figsize=(10,6))
sns.scatterplot(ridge.predict(X_train), resids)
ax.axhline(0)
ax.set_title("Residual vs Predicted Price")
ax.set_xlabel("Predicted SalePrice", fontsize=12, labelpad=20)
ax.set_ylabel("Residual / Difference\n from actual value", fontsize=12, labelpad=30, rotation=0);
```

```python
#Residuals vs. Predicted sale price for test data
resids = y_pred_ridge-y_test

fig,ax=plt.subplots(figsize=(10,6))
sns.scatterplot(y_pred_ridge, resids)
ax.axhline(0)
ax.set_title("Residual vs Predicted Price")
ax.set_xlabel("Predicted SalePrice", fontsize=12, labelpad=20)
ax.set_ylabel("Residual / Difference\n from actual value", fontsize=12, labelpad=30, rotation=0);
```

```python
ridgex = ridge_coefficients.sort_values(by="coefficient values", ascending=False)
top10 = ridgex.head(10).reset_index()
bottom10 = ridgex.tail(10).reset_index()
top10['index'] = [i for i in range(10)]
bottom10['index'] = [i for i in range(10)]
bottom10
topbtm10_coefficients =pd.merge(top10, bottom10, on=['index'])
```

```python
topbtm10_coefficients_2 =pd.concat((top10, bottom10), axis=0)
topbtm10_coefficients_2.sort_values(by='coefficient values', ascending=False, inplace=True)
topbtm10_coefficients_2
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data = topbtm10_coefficients_2, x = 'features', y = 'coefficient values')
plt.xticks(rotation = 60);
```

### 5. Regularizing the model: StandardScaler()
Used to prevent overfitting

```python
#StandardScaler()
ss = StandardScaler()
ss.fit(X_train)
X_train_sc = ss.transform(X_train)
X_test_sc = ss.transform(X_test)
```

```python
#Train
lr.fit(X_train_sc, y_train);
ridge.fit(X_train_sc, y_train);
lasso.fit(X_train_sc, y_train);

print(cross_val_score(lr, X_train_sc, y_train, cv=5).mean())
print(cross_val_score(ridge, X_train_sc, y_train, cv=5).mean())
print(cross_val_score(lasso, X_train_sc, y_train, cv=5).mean())
```

```python
#Test
lr.fit(X_test_sc, y_test);
ridge.fit(X_test_sc, y_test);
lasso.fit(X_test_sc, y_test);

print(cross_val_score(lr, X_test_sc, y_test, cv=5).mean())
print(cross_val_score(ridge, X_test_sc, y_test, cv=5).mean())
print(cross_val_score(lasso, X_test_sc, y_test, cv=5).mean())
```

```python
#Ridge is the best model so let's look at it further
ridge = Ridge(alpha=10)
ridge.fit(X_train_sc, y_train)
y_pred_ridge = ridge.predict(X_test_sc)
# R2
print(f"R2 score: {r2_score(y_test, y_pred_ridge)}")
# RMSE
print(f"RMSE score: {mean_squared_error(y_test, y_pred_ridge, squared=False)}")
```
